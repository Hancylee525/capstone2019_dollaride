### find the inter-zone travel with most trip counts

```python
df631 = df63.sort_values(by='TripCount', ascending=False)
# df63 = df63.drop(df63[df63['zone_x'] == df63['zone_y']].index, inplace = True)
df631['InterZone_Count'] = df631.apply(lambda x: x.zone_x in x.zone_y, axis=1)
# https://stackoverflow.com/questions/43855685/python-pandas-check-if-string-in-one-column-is-contained-in-string-of-another-c
df631 = df631[df631['InterZone_Count'] == False]
df631 = df631.drop(columns='InterZone_Count', axis=1)
df631[:5]
```

```python
def topDest(zone):
    df632 = df631[df631['zone_x'] == zone]
    df633 = df632.merge(TZ, left_on = 'zone_y', right_on = 'zone')
    df634 = df633.iloc[:10,:]
    gpd634 = gpd.GeoDataFrame(df634)
    gpd634.crs = {'init':'epsg:4326'}
    
    df635 = df632.merge(TZ, left_on = 'zone_x', right_on = 'zone').iloc[:1,:]
    gpd635 = gpd.GeoDataFrame(df635)
    gpd635.crs = {'init':'epsg:4326'}
    
    gpd634['coords'] = gpd634['geometry'].apply(lambda x: x.representative_point().coords[:])
    gpd634['coords'] = [coords[0] for coords in gpd634['coords']]
    gpd635['coords'] = gpd635['geometry'].apply(lambda x: x.representative_point().coords[:])
    gpd635['coords'] = [coords[0] for coords in gpd635['coords']]
    
    f, ax = plt.subplots(figsize=(20, 20))
    ax.get_xaxis().set_visible(True)
    ax.get_yaxis().set_visible(True)

    TZ.plot(alpha=1, linewidth=1, ax=ax, legend=True,
        k=1, color='white', edgecolor='black')

    gpd634.plot(alpha=0.75, linewidth=0.1, ax=ax, legend=True,
            cmap='Blues', column='TripCount')
    gpd635.plot(alpha=0.5, ax=ax, color='green')

    # , cmap = 'hot_r', column = 'INCOME_GINI'
    # df_3_b.plot(ax=ax, column = 'Cluster')

    for idx, row in gpd634.iterrows():
        plt.annotate(s=row['zone'], xy=row['coords'],
                 horizontalalignment='center')
    for idx, row in gpd635.iterrows():
        plt.annotate(s=row['zone'], xy=row['coords'],
                 horizontalalignment='center',color='red')    
    plt.title('Top Destinations for %s' % zone, fontsize = 20)

    plt.xlabel('Longitude', fontsize = 20)
    plt.ylabel('Latitude', fontsize = 20)
    plt.savefig(zone.replace("/", "_") + '.png')
    return df632[:5]
```

```python
def topDest(zone):

    import os
    import sys
    import io
    import numpy as np
    import pandas as pd
    import geopandas as gpd
    import urllib
    from matplotlib import pyplot as plt
    from fiona.crs import from_epsg
    import requests

    %matplotlib inline

    from shapely.geometry import LineString, Point
    from geopy.distance import great_circle
    from shapely.geometry import MultiPoint
    from shapely.ops import nearest_points
    import shapely
    
    # load original dataset
#     df8 = pd.read_csv('Trips of NYC.csv') 
    df8 = df1
    # subset based on choice
    # Employed:
    # df9 = df8[df8['rpworker'] == 1]
    # Unemployed:
    # df9 = df8[df8['rpworker'] == 2]

    # male:
    # df9 = df8[df8['rpsex'] == 1]
    # female:
    # df9 = df8[df8['rpsex'] == 2]

    # mode: Carpool 
#     df9 = df8[df8['pmode_r2'] == 1]

    # mode: public transit
#     df9 = df8[df8['pmode_r2'] == 2]

    # mode: taxi 
    # df9 = df8[df8['pmode_r2'] == 3]

    # mode: bike 
    df9 = df8[df8['pmode_r2'] == 4]

    # mode: walk 
    # df9 = df8[df8['pmode_r2'] == 5]

    # mode: citibike 
    # df9 = df8[df8['pmode_r2'] == 6]

    # mode: FHV 
#     df9 = df8[df8['pmode_r2'] == 7]

    # mode: auto 
#     df9 = df8[df8['pmode_r2'] == 8]    
    
    
    
    # df.sample()
    df10 = df9.sample(frac=0.1, random_state=1)    
    
    # load Taxi Zone
    TZ = gpd.read_file("NYC Taxi Zones/geo_export_726abfbe-1b7f-44b4-8cfe-fa474a5fbfa1.shp")
    TZ = TZ[(TZ['borough'] == 'Queens') | (TZ['borough'] == 'Brooklyn') | (TZ['borough'] == 'Bronx')] 
    
    # code Origin Coord
    df10['lonlat']=list(zip(df10.o_y,df10.o_x))
    # Create Point Geometry for based on lonlat column
    df10['geometry']=df10[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
    gpd10 = gpd.GeoDataFrame(df10)
    gpd10.crs = {'init':'epsg:4326'}    
    
    # code Destination Coord
    df11 = df10
    df11['lonlat']=list(zip(df10.d_y,df10.d_x))
    # Create Point Geometry for based on lonlat column
    df11['geometry']=df11[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
    gpd11 = gpd.GeoDataFrame(df11)
    gpd11.crs = {'init':'epsg:4326'}    
   
    # sjoin origin w/ TZ
    df12 = gpd.sjoin(gpd10, TZ)
    df121 = df12.reset_index()
    df121 = df121.rename(columns = {'index' : 'id_o'})
    # sjoin destination w/ TZ    
    df13 = gpd.sjoin(gpd11, TZ)
    df131 = df13.reset_index()
    df131 = df131.rename(columns = {'index' : 'id_d'})

    # merge sjoin origin and destination
    df14 = df121.merge(df131, left_on = 'id_o', right_on = 'id_d')
    df141 = df14[['id_o', 'geometry_x',  'borough_x',
            'objectid_x', 'shape_area_x', 'shape_leng_x', 'zone_x',  'geometry_y',
            'borough_y', 'objectid_y',
            'shape_area_y', 'shape_leng_y', 'zone_y']]
    
    # trip pair counts
    df15 = pd.DataFrame(df141.groupby(['zone_x', 'zone_y'])['id_o'].count().reset_index())
    df15 = df15.rename(columns = {'id_o':'TripCount'})   
    
    # inter-zone travel
    df631 = df15.sort_values(by='TripCount', ascending=False)
    # df63 = df63.drop(df63[df63['zone_x'] == df63['zone_y']].index, inplace = True)
    df631['InterZone_Count'] = df631.apply(lambda x: x.zone_x in x.zone_y, axis=1)
    # https://stackoverflow.com/questions/43855685/python-pandas-check-if-string-in-one-column-is-contained-in-string-of-another-c
    df631 = df631[df631['InterZone_Count'] == False]
    df631 = df631.drop(columns='InterZone_Count', axis=1)
    
    
    df632 = df631[df631['zone_x'] == zone]
    df633 = df632.merge(TZ, left_on = 'zone_y', right_on = 'zone')
    df634 = df633.iloc[:20,:]
    gpd634 = gpd.GeoDataFrame(df634)
    gpd634.crs = {'init':'epsg:4326'}
    
    df635 = df632.merge(TZ, left_on = 'zone_x', right_on = 'zone').iloc[:1,:]
    gpd635 = gpd.GeoDataFrame(df635)
    gpd635.crs = {'init':'epsg:4326'}
    
    gpd634['coords'] = gpd634['geometry'].apply(lambda x: x.representative_point().coords[:])
    gpd634['coords'] = [coords[0] for coords in gpd634['coords']]
    gpd635['coords'] = gpd635['geometry'].apply(lambda x: x.representative_point().coords[:])
    gpd635['coords'] = [coords[0] for coords in gpd635['coords']]
    
    f, ax = plt.subplots(figsize=(20, 20))
    ax.get_xaxis().set_visible(True)
    ax.get_yaxis().set_visible(True)

    TZ.plot(alpha=1, linewidth=1, ax=ax, legend=True,
        k=1, color='white', edgecolor='black')

    gpd634.plot(alpha=0.75, linewidth=0.1, ax=ax, legend=True,
            cmap='Blues', column='TripCount')
    gpd635.plot(alpha=0.5, ax=ax, color='green')

    # , cmap = 'hot_r', column = 'INCOME_GINI'
    # df_3_b.plot(ax=ax, column = 'Cluster')

    for idx, row in gpd634.iterrows():
        plt.annotate(s=row['zone'], xy=row['coords'],
                 horizontalalignment='center')
    for idx, row in gpd635.iterrows():
        plt.annotate(s=row['zone'], xy=row['coords'],
                 horizontalalignment='center',color='red')    
    plt.title('Top Destinations for %s' % zone, fontsize = 20)

    plt.xlabel('Longitude', fontsize = 20)
    plt.ylabel('Latitude', fontsize = 20)
    plt.savefig(zone.replace("/", "_") + '.png')
    return df632[:5]
```
