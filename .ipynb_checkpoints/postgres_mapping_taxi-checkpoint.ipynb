```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2
import geopandas as gpd

import pass_kp

import matplotlib.pyplot as plt
```

## Set the Query
---

```python
sql = "select pulocationid, dolocationid, count (*) \
    from taxi_large \
    where date_part('hour', pickup_datetime) between '{}' and '{}' and date_part('dow', pickup_datetime) in ('1','2','3','4','5') \
    group by pulocationid, dolocationid;".format('06', '22')

dat = sqlio.read_sql_query(sql, pass_kp.conn)
```

```python
import tx_kp
df = tx_kp.txtx(dat)
```

```python
tx_kp.tz.LocationID = tx_kp.tz.LocationID.astype(str)
```

```python
df_ = df.query('zone_x == zone_y')\
    .drop(['zone_x','borough_x','geometry_x','zone_y','borough_y','geometry_y'], axis=1)\
    .merge(tx_kp.tz, left_on='pulocationid', right_on='LocationID')
```

```python
df_ = gpd.GeoDataFrame(df_, geometry=df_.geometry_y)
```
